```python
import xml.etree.ElementTree as ET
import os

xml_dir = 'C:/Users/saira/project/images_new'

labels_dict = {'filepath': [], 'xmin': [], 'xmax': [], 'ymin': [], 'ymax': []}

# Iterate through all XML files in the directory
for filename in os.listdir(xml_dir):
    if filename.endswith(".xml"):
        xml_file = os.path.join(xml_dir, filename)
        
        # Parse the XML file
        tree = ET.parse(xml_file)
        root = tree.getroot()
        
        # Iterate through all 'object' elements in the XML
        for object_elem in root.findall('object'):
            labels_info = object_elem.find('bndbox')
            xmin = int(labels_info.find('xmin').text)
            xmax = int(labels_info.find('xmax').text)
            ymin = int(labels_info.find('ymin').text)
            ymax = int(labels_info.find('ymax').text)

            labels_dict['filepath'].append(xml_file)
            labels_dict['xmin'].append(xmin)
            labels_dict['xmax'].append(xmax)
            labels_dict['ymin'].append(ymin)
            labels_dict['ymax'].append(ymax)

# Now labels_dict contains all the extracted data
print(labels_dict)
```

```python
df = pd.DataFrame(labels_dict)
df.head(3)
```

```python
df.to_csv('labels.csv', index=False)
```
