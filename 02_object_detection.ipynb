## Data Preprocessing

```python
from sklearn.model_selection import train_test_split
from tensorflow.keras.preprocessing.image import load_img, img_to_array
```

```python
labels = df.iloc[:,1:].values
```

```python
len(image_path)
```

```python
data= []
output=[]

for ind in range(len(image_path)):
    image = image_path[ind]
    img_arr = cv2.imread(image)
    if img_arr is None:
        continue
    h, w, d = img_arr.shape

   # preprocessing
    load_image = cv2.resize(img_arr, (224, 224))
    load_image_arr = np.array(load_image)
    norm_load_image_arr = load_image_arr / 255.0  # normalization

    # normalization to labels
    xmin, xmax, ymin, ymax = labels[ind]
    nxmin, nxmax = xmin / w, xmax / w
    nymin, nymax = ymin / h, ymax / h
    labels_norm = (nxmin, nxmax, nymin, nymax)  # normalization output

    # append
    data.append(norm_load_image_arr)
    output.append(labels_norm)
```

```python
labels_norm
```

```python
load_image_arr
```

```python
X = np.array(data, dtype=np.float32)
y = np.array(output, dtype=np.float32)
```

```python
X.shape, y.shape
```

```python
x_train, x_test, y_train, y_test = train_test_split(X,y,test_size=0.2,random_state=0)
x_train.shape, x_test.shape, y_train.shape, y_test.shape
```

### Deep Learning Model

```python
#transfer learning
```

```python
from tensorflow.keras.applications import InceptionResNetV2
from tensorflow.keras.layers import Input, Flatten, Dense, Dropout, BatchNormalization
from tensorflow.keras.models import Model

# Load InceptionResNetV2 with pre-trained weights
inception_resnet = InceptionResNetV2(weights="imagenet", include_top=False,
                                      input_tensor=Input(shape=(224, 224, 3)))
inception_resnet.trainable = False

# Create the head of the model
headmodel = inception_resnet.output
headmodel = Flatten()(headmodel)
headmodel = Dense(500, activation="relu")(headmodel)
headmodel = BatchNormalization()(headmodel)
headmodel = Dropout(0.3)(headmodel)  # Add dropout
headmodel = Dense(250, activation="relu")(headmodel)
headmodel = BatchNormalization()(headmodel)
headmodel = Dropout(0.3)(headmodel)  # Add dropout
headmodel = Dense(4, activation='sigmoid')(headmodel)

```

```python
# Create the final model
model = Model(inputs=inception_resnet.input, outputs=headmodel)
```

```python
# Compile the model
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
```

```python
model.summary()
```

```python
from tensorflow.keras.callbacks import EarlyStopping

early_stopping = EarlyStopping(monitor='val_accuracy', patience=20, verbose=1, restore_best_weights=True)

```

```python
history = model.fit(x=x_train, y=y_train, batch_size=32, epochs=100,
                    validation_data=(x_test, y_test), callbacks=[early_stopping])

best_epoch = history.history['val_accuracy'].index(max(history.history['val_accuracy']))
best_accuracy = history.history['val_accuracy'][best_epoch]


print(f"Best validation accuracy of {best_accuracy} achieved at epoch {best_epoch + 1}")
```
